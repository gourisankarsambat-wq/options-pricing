**ORNSTEIN-UHLENBECK PROCESS**

It is a Stochastic process used to model mean-reverting quantitities

(eg:interest rates,Brownian particle with friction)

Defined by SDE:

            dX(t)​=θ(μ−X(t)​)dt+σdWt​

X(t)-OU process at time t

X(0)-OU process at time t=0

θ>0 — speed of mean reversion


μ — the long-run level the process is pulled toward


σ — volatility

Solution is given by:

$$X_t = X_0 e^{-\theta t} + \mu (1 - e^{-\theta t}) + \sigma \int_{0}^{t} e^{-\theta(t-s)} dW_s$$

**MEAN REVERSION**

This is the most important property of OU process.

The drift term actively pulls $X_t$back toward $\mu$
.

If $X_t > \mu$, the drift ($\mu - X_t$)
 is negative, pulling it down;

  if
$X_t​<μ$, the drift($\mu-X_t$) is positive, pulling it up.

**Stationary distribution**

as t→∞t

$$X_t→ \mathcal{N}\!\left(\mu, \frac{\sigma^2}{2\theta}\right)​$$
**Major difference from GBM:**

OU settles into a fixed long-run distribution, while GBM's variance grows without bound over time.

**Euler–Maruyama discretization**

$$X_(n+1)​=X_n​+θ(μ−X_n​)Δt+σΔW_n$$​

```python
import numpy as np
import matplotlib.pyplot as plt

#Parameters
X0=0.2
theta=0.7
mu=0.5
sigma=0.2
T=5              #Year
N=252*5
dt=T/N

X=np.zeros(N+1)
X[0]=X0
#getting OU by euler discretisation
for i in range(N):
  dW=np.random.normal(0,dt)
  X[i+1]=X[i]+ theta*(mu-X[i])*dt +sigma*dW

#Plotting
t=np.linspace(0,T,N+1)
plt.plot(t,X)
plt.axhline(y=mu, color='red', linestyle='--',label='Long-run mean (μ)' )
plt.legend()
plt.xlabel("Time")
plt.ylabel("X(t)")
plt.title("OU process via Euler-Maruyama")
plt.show()
```
