**GEOMETRIC BROWNIAN MOTION(GBM)**

Standard model used in stock prices

Defined by SDE:

            dX(t)​=μX(t)​dt+σX(t)​dWt​
Solution is given by

             X(t)​=X(0​)exp[(μ-(σ^2)​/2)t+σWt​]

X(t): Price of stock at time t

X(0): Price of stock at time t=0

μ: Drift term(Expected annualised return of asset)

σ: Diffusion term(Volatility)

Wt: Brownian motion


**Key Properties:**

1.**X(t) is always positive:**

Fixes a major flow in standard brownian motion as stock prices cant be negative

2.**Log returns are normally distributed**

This makes returns easier to model


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
X0 = 100       # initial value(Arbitarily given unlike standard brownian motion)
mu = 0.08       # drift
sigma = 0.2     # volatility
T = 1           # time horizon (years)
N = 252       # number of steps(Trading days in an year)
dt = T / N      # step size

# Euler-Maruyama simulation
X = np.zeros(N + 1)
X[0] = X0
for i in range(N):
    dW = np.random.normal(0,dt)                # Brownian increment
    X[i + 1] = X[i] + mu * X[i] * dt + sigma * X[i] * dW

# Plot
t = np.linspace(0, T, N + 1)
plt.plot(t, X)
plt.xlabel("Time")
plt.ylabel("X(t)")
plt.title("GBM via Euler-Maruyama")
plt.show()
```
